### Explanation of the data variables

- trial_index = number of trials
- block_index = number of blocks (8 blocks in total, block -1 is practice block)
- block_trial = number of trials within blocks
- block_mu = mean orientation of the prior within a block (same first 6 blocks)
- block_sigma = standard deviation of prior within a block (always 20°)
- trial_mu_deviation = deviation mean orientation from mean of prior
- trial_mu = mean orientation of distribution from with the directions of moving dots are sampled per trial
- trial_sigma = variability of moving dots per trial (10 or 40°)
- trial_mean = actual mean orientation of moving dots per trial
- initial_ori = initial orientation of the fishing rod
- final_ori = response
- rt


```python
#create factor variables for means of priors

# prior_mean is factor variable for mean of prior
df['prior_mean'] = np.where(df['block_mu'] == 22, 22, 68)

# Session variable indicates whether the first prior was the one with 22 mean or 68 mean.
df['session'] = np.where(((df.block_mu == 22) & (df.block_index.isin([0,1,2,3,4,5]))) | ((df.block_mu == 68) & (df.block_index.isin([6,7]))) , "22_first", "68_first")

#Create factor variable for blocks of 100 trials
def blocks(series):
    if series < 2:
        return "1"
    elif 1 < series < 4:
        return "2"
    elif 3 < series < 6:
        return "3"
    elif series > 5:
        return "4"

df['tr_100'] = df['block_index'].apply(blocks)

#Create factor variable to indicate whether it is the first or the second prior
df['prior'] = np.where((df.block_index < 6), "prior 1", "prior 2") #prior 1 or prior 2
```

## Adding key computed variables

```python

#distance_prior_response is the final chosen direction minus mean of the prior in that block; distance_prior is the presented motion direction minu the mean of the prior in that block. 
df['distance_prior_response'] = df.final - df.block_mu
df['distance_prior'] = df.trial_mu - df.block_mu 
df['distance_prior_response'] = np.where((df.distance_prior_response > 180), df.distance_prior_response - 360, df.distance_prior_response)
df['distance_prior_response'] = np.where((df.distance_prior_response < -180), df.distance_prior_response + 360, df.distance_prior_response)
df['bias']= df.distance_prior_response - df.distance_prior
df['bias'] = np.where((df.bias < -180), df.bias + 360, df.bias)
df['bias'] = np.where((df.bias > 180), df.bias - 360, df.bias)
#slope like in Vilares (see supplementary)
df['slope']=df.distance_prior_response/(df.trial_mu - df.block_mu)
#Instantaneous slope like in Vilares (see supplementary)
df['arc_slope'] = (np.arctan(-0.5 + df.slope)/np.pi)+0.5
#rt normalised by distance between chosen orientation and initial orientation of the presented fishing rod
df['rt_norm']=df.rt/(abs(df.final_ori-df.initial_ori))
#Convert the responses of sessions with 22 degree prior first into 68 degrees first by inverting the sign.
df['distance_prior_response_conv'] = np.where(df.session =="22_first", df.distance_prior_response*(-1), df.distance_prior_response)
#same for the distance of the presented directions from the priors in the 22° first sessions.
df['distance_prior_conv'] = np.where(df.session =="22_first", df.distance_prior*(-1), df.distance_prior)
#same for bias.
df['bias_conv'] = np.where(df.session =="22_first", df.bias*(-1), df.bias)
```

# Compare (rms) errors between conditions

```python
def fu(df):
    d = {}
    d['group']= df['group'].iloc[0]
    d['rms_error'] = np.sqrt(np.mean(df.distance_prior_response**2))
    d['abs_error'] = np.mean(np.abs(df.distance_prior_response))
    return pd.Series(d)

# make df for grouped error data
error = df.groupby(['id','trial_sigma']).apply(fu).reset_index()

for var in ["rms_error","abs_error"]:
    # Initialize the figure
    f, ax = plt.subplots()
    sns.despine(bottom=False, left=False)

    # Show each observation with a scatterplot
    sns.stripplot(x="trial_sigma", y=var, hue="group",
                  data=error, dodge=True, jitter=True,
                  alpha=.4, zorder=1)

    # Show the conditional means
    sns.pointplot(x="trial_sigma", y=var, hue="group",
                  data=error, dodge=.532, join=False, palette="dark",
                  markers="d", scale=.75, ci=None)

    # Improve the legend 
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles[2:], labels[2:],
              handletextpad=0, columnspacing=1,
              loc="upper left", ncol=2, frameon=True)
```

```python
# errors per block
errorblock = df.groupby(['id','tr_100','trial_sigma']).apply(fu).reset_index()   

g = sns.FacetGrid(errorblock, col="trial_sigma", hue="group", height=6, aspect=.7)
g.map(sns.stripplot, "tr_100", "abs_error", alpha=.3);
g.map(sns.pointplot, "tr_100", "abs_error", jitter=.8);

g.add_legend();

    
```

```python
results = smf.ols("rms_error ~ trial_sigma*group + trial_sigma*tr_100", data=errorblock).fit()
print(results.summary())

md = smf.mixedlm("rms_error ~ trial_sigma*group + group*tr_100", errorblock, groups=errorblock["id"]).fit()

md.summary()
```

# Fit individual linear regressions to compare slopes

```python
#Function to do a linear regression for each person's data and give the slope, intercept and correlation
def est_slope(df): 
    d = {}
    d['group']= df['group'].iloc[0]
    d['session']= df['session'].iloc[0]
    d['slopeV'] = df['slope'].mean()
    d['arc_slope'] = df['arc_slope'].mean()
    lm = pg.linear_regression(df.distance_prior_conv, df.distance_prior_response_conv)
    d['intercept'] = lm.coef[0]
    d['slope'] = lm.coef[1]
    d['correlation'] = pg.corr(df.distance_prior_response_conv,df.distance_prior_conv).r[0]
    return pd.Series(d)

#Slope per person, group, trial_sigma and prior
df_est_cor_slope = df.groupby(['id','trial_sigma', 'prior']).apply(est_slope).reset_index()

zscore = lambda x: (x - x.mean()) / x.std()
df_est_cor_slope['slope_z']= df_est_cor_slope.groupby(['id'])['slope'].transform(zscore)
df_est_cor_slope['corr_z']= df_est_cor_slope.groupby(['id'])['correlation'].transform(zscore)
df_est_cor_slope['intercept_z']= df_est_cor_slope.groupby(['id'])['intercept'].transform(zscore)
```

```python
sns.regplot(x="slope", y="arc_slope", data=df_est_cor_slope);
```

```python

for var in ["slope","intercept","correlation"]:
    g = sns.FacetGrid(df_est_cor_slope, col="trial_sigma", hue="group", height=6, aspect=.7)
    g.map(sns.stripplot, "prior", var, alpha=.3);
    g.map(sns.pointplot, "prior", var, jitter=.8);
    g.add_legend();
```

```python
# Is it necessary to do mixed model if standardized slope (based on id)?
results = smf.ols('slope_z ~ group  * trial_sigma * prior', data=df_est_cor_slope).fit()
print(results.summary())

#the R lmer model also gives a significant interaction between prior and group (and similar effect of sigma & prior)        
#lmer_slope <- lmer(slope_z ~ group  * trial_sigma* prior + (trial_sigma|id)+ (prior|id), REML=F, data = df_est_cor_slope)
#summary(lmer_slope)

results = smf.ols('intercept_z ~ group  * trial_sigma * prior', data=df_est_cor_slope).fit()
print(results.summary())


from bambi import Model

# Assume we already have our data loaded
model = Model(df_est_cor_slope)
results = model.fit('slope_z ~ 1 + group * trial_sigma * prior', random=['trial_sigma|id', 'prior|id', '1|id'], samples=5000, chains=2)
results[1000:].plot()
results[1000:].summary()
```

# Comparing instantaneous slope of last 20 trials before and 20 trials after shift in prior

```python

#Instantaneous slope of last 20 trials before and 20 trials after shift in prior (if distance prior is 0, slope is 0 as well)
slopes_shift = (df.query('distance_prior != 0')
                  .query('(trial_index>287 & trial_index<307) | (trial_index<327 & trial_index>306)')
                  .groupby(['group','id','block_index','trial_sigma'])
                  .arc_slope.mean().reset_index()
                  .pivot_table(index=['id','group','trial_sigma'],columns='block_index',values='arc_slope').reset_index()
                  .rename(columns={5: "last_20", 6: "first_20"})
                  .assign(diff_slope=lambda x: x.first_20 - x.last_20))




# Initialize the figure
f, ax = plt.subplots()
sns.despine(bottom=False, left=False)

# Show each observation with a scatterplot
sns.stripplot(x='trial_sigma', y="diff_slope", hue="group", data=slopes_shift, dodge=True, jitter=True,alpha=.4, zorder=1)

# Show the conditional means
sns.pointplot(x="trial_sigma", y="diff_slope", hue="group",
              data=slopes_shift, dodge=.532, join=False, palette="dark",
              markers="d", scale=.75)

# Improve the legend 
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[2:], labels[2:],
          handletextpad=0, columnspacing=1,
          loc="upper right", ncol=2, frameon=True);

md = smf.mixedlm("diff_slope ~ group*trial_sigma", slopes_shift, groups=slopes_shift["id"]).fit()
print(md.summary())
#results = smf.ols('diff_slope ~ group  * trial_sigma', data=slopes_shift).fit()
#print(results.summary())
```
